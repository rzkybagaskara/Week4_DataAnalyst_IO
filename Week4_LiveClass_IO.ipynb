#### Checking for Outliers

```python
#labels
unit_price = df['Unit_price']
qty = df['Quantity']
tax = df['Tax']
total = df['Total']
cogs = df['cogs']
gross_income = df['gross_income']
ratings = df['Rating']
```

Boxplot for detecting outliers

```python
#creating subplots
fig, axes = plt.subplots(1, 7, figsize = (25,8))
fig.suptitle('Outlier check using Boxplot')

sns.boxplot(unit_price, ax=axes[0])
axes[0].set_title('Unit_Price')

sns.boxplot(qty, ax=axes[1])
axes[1].set_title('Quantity')

sns.boxplot(tax, ax=axes[2])
axes[2].set_title('Tax')

sns.boxplot(total, ax=axes[3])
axes[3].set_title('Total')

sns.boxplot(cogs, ax=axes[4])
axes[4].set_title('COGS')

sns.boxplot(gross_income, ax=axes[5])
axes[5].set_title('Gross_Income')

sns.boxplot(ratings, ax=axes[6])
axes[6].set_title('Ratings')
```

Data Distribution

```python
#creating subplots
fig, axes = plt.subplots(1, 7, figsize = (25,8))
fig.suptitle('Outlier check using Boxplot')

sns.histplot(unit_price, kde=True, ax=axes[0])
axes[0].set_title('Unit_Price')

sns.histplot(qty, kde=True, ax=axes[1])
axes[1].set_title('Quantity')

sns.histplot(tax, kde=True, ax=axes[2])
axes[2].set_title('Tax')

sns.histplot(total, kde=True, ax=axes[3])
axes[3].set_title('Total')

sns.histplot(cogs, kde=True, ax=axes[4])
axes[4].set_title('COGS')

sns.histplot(gross_income, kde=True, ax=axes[5])
axes[5].set_title('Gross_Income')

sns.histplot(ratings, kde=True, ax=axes[6])
axes[6].set_title('Ratings')
```

#### Relationship Analysis

##### Gross Income

COGS <=> Gross_Income

```python
cogs_gross_income = df[['cogs', 'gross_income']]

r = scipy.stats.pearsonr(x=df['cogs'], y=df['gross_income'])[0]

sns.scatterplot(data=cogs_gross_income, x='cogs', y='gross_income')
plt.title('Relatioship between COGS and Gross Income')

plt.text(5,30, 'r = ' + str(round(r, 4)), fontsize=15)

plt.show()
```

Unit_Price <=> Gross_Income

```python
unitprice_gross_income = df[['Unit_price', 'gross_income']]

r = scipy.stats.pearsonr(x=df['Unit_price'], y=df['gross_income'])[0]

sns.scatterplot(data=unitprice_gross_income, x='Unit_price', y='gross_income')
plt.title('Relatioship between Unit Price and Gross Income')

plt.text(5,30, 'r = ' + str(round(r, 4)), fontsize=15)

plt.show()
```

Quantity <=> Gross_Income

```python
qty_gross_income = df[['Quantity', 'gross_income']]

r = scipy.stats.pearsonr(x=df['Quantity'], y=df['gross_income'])[0]

sns.regplot(data=qty_gross_income, x='Quantity', y='gross_income')
plt.title('Relatioship between Quantity and Gross Income')

plt.text(5,30, 'r = ' + str(round(r, 4)), fontsize=15)

plt.show()
```

Tax <=> Gross_Income

```python
tax_gross_income = df[['Tax', 'gross_income']]

r = scipy.stats.pearsonr(x=df['Tax'], y=df['gross_income'])[0]

sns.scatterplot(data=tax_gross_income, x='Tax', y='gross_income')
plt.title('Relatioship between Tax and Gross Income')

plt.text(5,30, 'r = ' + str(round(r, 4)), fontsize=15)

plt.show()
```

Total <> Gross_Income 

```python
total_gross_income = df[['Total', 'gross_income']]

r = scipy.stats.pearsonr(x=df['Total'], y=df['gross_income'])[0]

sns.regplot(data=total_gross_income, x='Total', y='gross_income')
plt.title('Relationship between Total and Gross Income')

plt.text(5,30, 'r = ' + str(round(r, 4)), fontsize=15)

plt.show()
```

**Penyebab gross_income tinggi karena Totalnya tinggi, total merupakan harga barang + pajak (Tax) 5%.**

**Semakin tinggi harga barang (Total) yang terjual, maka semakin tinggi juga total keuntungan kotornya.**

#### Sales Breakdown (Transactions)

Date

```python
sales_through_dates = df.groupby('DD').count().reset_index()

sns.lineplot(data=sales_through_dates, x='DD', y='Invoice_ID', sort=True)
plt.title('Sales through Dates')
plt.xlabel('Date')
plt.ylabel('Transactions')
plt.show()
```

Month

```python
sales_through_month = df.groupby('MM').count().reset_index()

sns.lineplot(data=sales_through_month, x='MM', y='Invoice_ID', sort=True)
plt.title('Sales through Month')
plt.xlabel('Month')
plt.ylabel('Transactions')
plt.show()
```

Branch

```python
sales_per_branch = df.groupby('Branch', as_index=False).count()

ax = sns.barplot(data=sales_per_branch, x='Branch', y='Invoice_ID', 
            order=sales_per_branch.sort_values('Invoice_ID', ascending=False).Branch)

#data labels
ax.bar_label(ax.containers[0], fontsize=11)

plt.title('Sales per Branch')
plt.xlabel('Branch')
plt.ylabel('Transactions')
plt.show()
```

City

Gender
